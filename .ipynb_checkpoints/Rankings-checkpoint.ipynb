```python
import pandas as pd
```

```python
espn = pd.DataFrame.from_csv('espn2.csv')
cbs = pd.DataFrame.from_csv('cbs.csv')
snews = pd.DataFrame.from_csv('sportnews.csv')
```

```python
espn
```

```python
cbs
```

```python
snews
```

```python
merged = pd.merge(pd.merge(espn, cbs, on='Player', how='outer'), snews, on='Player', how='outer')
```

```python
merged.columns
```

```python
merged = merged.set_index('Player', drop=True)
```

```python
merged.columns = ['ESPN', 'CBS', 'SportingNews']
```

```python
merged
```

```python
merged['Mean'] = merged.mean(axis=1)
```

```python
merged
```

```python
merged['Count'] = merged.count(axis=1)
```

```python
merged['Count'] = merged['Count'] - 1
```

```python
merged = merged.sort_values('Mean')
```

```python
merged
```

```python
merged.to_csv('ECSjoin.csv')
```
